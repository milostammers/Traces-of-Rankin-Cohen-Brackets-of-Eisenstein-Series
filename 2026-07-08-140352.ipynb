```sage
from sympy.functions.combinatorial.numbers import kronecker_symbol
import math
import numpy
from sympy import divisors, factorint
import sage.lfunctions.lcalc


# Computes (-I)^k
def Ipow(k):
    if k%4==0:
        return 1
    if k%4==1:
        return -I
    if k%4==2:
        return -1
    return I

# Determines if n is an odd fundamental discriminant
def isFundDisc(n):
    if n%4 != 1:
        return False
    primepowers = factorint(n).values()
    return all(exp == 1 for exp in primepowers)

# Computes C_{2, a_1, a_2}/(k*(k+1)) for weight k
def C(k, a_1, a_2):
    return (0.5)*(a_1*a_1+a_2*a_2)-(k+1)*(a_1)*(a_2)/(k)

# Computes sigma_{k-1, D_1, D_2}(n)*(2*pi*e/((k-1)|D|)^{k-1}
def sigma(k, D_1, D_2, D, n):
    sum = 0
    for d_1 in divisors(n):
        d_2 = n//d_1
        sum += kronecker_symbol(D_1, d_1)*kronecker_symbol(D_2, d_2)*math.pow(d_1*2*math.pi*math.e/((k-1)*abs(D)), k-1)
    return sum

# Computes f'_{D, k, 2}(1)*(2*pi*e/(k-1)|D|)^{k-1}
def fprime(k, D):
    sum = 0
    Ddiv = divisors(abs(D))
    for absD_2 in Ddiv:
        if absD_2%4 == 1:
            D_2 = absD_2
        else:
            D_2 = -absD_2
        D_1 = D//D_2
        for a_1 in range(1,absD_2):
            a_2 = absD_2 - a_1
            sum += kronecker_symbol(D_2, -1)*math.pow(absD_2, -2)*sigma(k, D_1, D_2, D, (a_1*a_2))*C(k, a_1, a_2)
    return sum

# Checks non-vanishing of the first Fourier coefficient for k <= min((A-4)/2, 2*pi*e*|D|)
def nonVanFC(A):
    for D in range(-(int)((A-4)/(4*math.pi*math.e))-1, (int)((A-4)/(4*math.pi*math.e)+1)):
        if not isFundDisc(D):
            continue
        print("D =", D)
        print("-"*50)
        print("")
        for k in range(4, (int)(2*math.pi*math.e*abs(D))):
            if math.pow(-1, k)*(abs(D)//D) == -1:
                continue
            fp = fprime(k,D)
            if D != 1:
                chi = kronecker_character(D)
                f0 = math.sqrt(2*math.pi*(k-1))*D*math.pow(D//abs(D), k-1)*(lcalc.twist_values(k, D, D)[0][1])/(Ipow(k)*2*math.pi*chi.gauss_sum_numerical())
            else:
                f0 = math.sqrt(2*math.pi*(k-1))*(lcalc.value(k))/(Ipow(k)*2*math.pi)
            ratio = -fp/f0
            print("k =", k)
            print(ratio)
            print("")
            err = numpy.log(ratio)*(k-1)
            if 0 < err and err < 0.2:
                print("!"*100)
                print("Potential Failure: D =", str(D), "k =", str(k))
                print("Ratio =", str(ratio))
                print("!"*100)
                print("")
    
```
